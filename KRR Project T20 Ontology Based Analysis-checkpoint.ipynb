# Creating RDF file

```python
from rdflib import Graph, Literal, Namespace, RDF, URIRef
from rdflib.namespace import XSD


onto = Namespace("http://www.semanticweb.org/user/ontologies/2024/12#")

# Instantiate a new RDF graph
g = Graph()

# Define namespaces
g.bind("owl", "http://www.w3.org/2002/07/owl#")
g.bind("xsd", "http://www.w3.org/2001/XMLSchema#")
g.bind("", onto)

# Function to generate RDF triples for each row of innings data
# def create_rdf_triples(data, innings):
#     for index, row in data.iterrows():
#         match_id = row['matchId']
#         balls = row['balls']
#         runs = row['runs']
#         toss_winner = row['tossWinner']
#         batting_team = row['battingTeam']
#         bowling_team = row['bowlingTeam']
#         player_out = row['player_out']
#         city = row['city']
#         venue = row['venue']

#         # Create unique URIs for match and innings
#         match_uri = onto[f"Match_{innings}_{match_id}"]
#         inning_uri = onto[f"{innings}_Innings_T20"]

#         # Add triples to the graph
#         g.add((match_uri, RDF.type, inning_uri))
#         g.add((match_uri, onto.hasBalls, Literal(balls, datatype=XSD.float)))
#         g.add((match_uri, onto.hasRuns, Literal(runs, datatype=XSD.int)))
#         g.add((match_uri, onto.hasTossWinner, Literal(toss_winner)))
#         g.add((match_uri, onto.hasBattingTeam, Literal(batting_team)))
#         g.add((match_uri, onto.hasBowlingTeam, Literal(bowling_team)))
#         g.add((match_uri, onto.hasPlayerOut, Literal(player_out, datatype=XSD.int)))
#         g.add((match_uri, onto.hasCity, Literal(city)))
#         g.add((match_uri, onto.hasVenue, Literal(venue)))

def create_rdf_triples(data, innings):
    def sanitize_name(name):
        if isinstance(name, str):  # Check if the value is a string
            return name.replace(" ", "_").replace("&", "and").replace("/", "_")
        else:
            return "Unknown_Player"  # Default for non-string values

    for index, row in data.iterrows():
        match_id = row['matchId']
        balls = row['balls']
        runs = row['runs']
        toss_winner = sanitize_name(row['tossWinner'])
        batting_team = sanitize_name(row['battingTeam'])
        bowling_team = sanitize_name(row['bowlingTeam'])
        player_out = sanitize_name(row['player_out']) if not pd.isnull(row['player_out']) else "Unknown_Player"
        city = row['city']
        venue = row['venue']

        match_uri = onto[f"Match_{innings}_{match_id}"]
        inning_uri = onto[f"{innings}_Innings_T20"]

        g.add((match_uri, RDF.type, inning_uri))
        g.add((match_uri, onto.hasBalls, Literal(balls, datatype=XSD.float)))
        g.add((match_uri, onto.hasRuns, Literal(runs, datatype=XSD.int)))
        g.add((match_uri, onto.hasTossWinner, URIRef(onto[toss_winner])))
        g.add((match_uri, onto.hasBattingTeam, URIRef(onto[batting_team])))
        g.add((match_uri, onto.hasBowlingTeam, URIRef(onto[bowling_team])))
        g.add((match_uri, onto.hasPlayerOut, URIRef(onto[player_out])))
        g.add((match_uri, onto.hasCity, Literal(city)))
        g.add((match_uri, onto.hasVenue, Literal(venue)))

        g.add((URIRef(onto[batting_team]), onto.hasBalls, Literal(balls, datatype=XSD.float)))
        g.add((URIRef(onto[batting_team]), onto.hasRuns, Literal(runs, datatype=XSD.int)))
        g.add((URIRef(onto[batting_team]), onto.hasTossWinner, URIRef(onto[toss_winner])))
        g.add((URIRef(onto[batting_team]), onto.hasCity, Literal(city)))
        g.add((URIRef(onto[batting_team]), onto.hasVenue, Literal(venue)))
        
        g.add((URIRef(onto[bowling_team]), onto.hasBalls, Literal(balls, datatype=XSD.float)))
        g.add((URIRef(onto[bowling_team]), onto.hasRuns, Literal(runs, datatype=XSD.int)))
        g.add((URIRef(onto[bowling_team]), onto.hasTossWinner, URIRef(onto[toss_winner])))
        g.add((URIRef(onto[bowling_team]), onto.hasCity, Literal(city)))
        g.add((URIRef(onto[bowling_team]), onto.hasVenue, Literal(venue)))

        g.add((URIRef(onto[batting_team]), RDF.type, onto.Team))
        g.add((URIRef(onto[bowling_team]), RDF.type, onto.Team))
        g.add((URIRef(onto[player_out]), RDF.type, onto.Player))

        
# Assuming you have parsed your data into pandas DataFrames
innings1_data = pd.read_csv('1st Innings T20.csv')
innings2_data = pd.read_csv('2nd Innings T20.csv')

# Create RDF triples for each innings
create_rdf_triples(innings1_data, "1st")
create_rdf_triples(innings2_data, "2nd")

g.serialize(destination='latest_data.rdf', format='xml')
```

# Classes

```python
g = Graph()
g.parse("latest_data.rdf", format='xml')

query_classes = """
SELECT DISTINCT ?class 
WHERE {
    ?s a ?class .   # a means rdf type
    FILTER(isURI(?class))
}
"""
classes = g.query(query_classes)

print("Classes:")
for cls in classes:
    print(cls)
```

# Relations

```python
# Get all relations (properties)
query_relations = """
SELECT DISTINCT ?property 
WHERE {
    ?s ?property ?o .
    FILTER(isURI(?property))
}
"""
relations = g.query(query_relations)

print("\nRelations (Properties):")
for rel in relations:
    print(rel)
```

# Relations of the Team Class

```python
# Define the URI for the Team class
team_class = "http://www.semanticweb.org/user/ontologies/2024/12#Team"

# SPARQL query to get all properties specifically for Team class instances
query = """
    SELECT DISTINCT ?property
    WHERE {
        ?team rdf:type <""" + team_class + """> .      # Match instances of the Team class
        ?team ?property ?value .                        # Find all properties for those instances
        FILTER (?property != rdf:type)                   # Exclude rdf:type from the result
    }
"""

# Execute the query
results = g.query(query)

# Print the properties
print("Properties of the Team class (excluding rdf:type):")
for result in results:
    print(result[0])
```

# Instances and their corresponding classes

```python
query_instances_and_classes = """
SELECT DISTINCT ?instance ?relation ?class
WHERE {
    ?instance ?relation ?class .   # Fetch instances, their relations, and corresponding classes
    FILTER(isURI(?class))  # Ensure that only URIs are returned for the classes (excluding literals)
}
"""

# Execute the query
instances_and_classes = g.query(query_instances_and_classes)

# Print the results
print("Instances and their corresponding classes with relations:")
for row in instances_and_classes:
    instance = row[0]  # instance URI
    relation = row[1]  # relation (predicate) URI
    class_uri = row[2]  # class URI
    print(f"Instance: {instance}, Relation: {relation}, Class: {class_uri}")
```

# Player Performance

```python
from rdflib import Graph, Namespace, RDF

# Load RDF file into the graph
g = Graph()
g.parse("latest_data.rdf")

# Query to find the top run-scorers
query_top_run_scorers = """
    PREFIX onto: <http://www.semanticweb.org/user/ontologies/2024/12#>
    SELECT ?player (SUM(?runs) AS ?total_runs)
    WHERE {
        ?match onto:hasPlayerOut ?player_out .
        ?match onto:hasRuns ?runs .
        FILTER(?player_out = 0)  # Assuming 0 denotes player not out
    }
    GROUP BY ?player
    ORDER BY DESC(?total_runs)
    LIMIT 10
"""

# Execute the query
result = g.query(query_top_run_scorers)

# Display top run-scorers
for row in result:
    print(f"Player: {row['player']} - Total Runs: {row['total_runs']}")
```

# Average Runs by Player

```python
query_avg_runs_by_player = """
    PREFIX onto: <http://www.semanticweb.org/user/ontologies/2024/12#>
    SELECT ?player (AVG(?runs) AS ?avg_runs)
    WHERE {
        ?match onto:hasPlayerOut ?player_out .
        ?match onto:hasRuns ?runs .
        ?match onto:hasBattingTeam ?team .
        ?match onto:hasBowlingTeam ?opponent_team .
        ?match onto:hasPlayerOut ?player .
    }
    GROUP BY ?player
    ORDER BY DESC(?avg_runs)
    LIMIT 10
"""

# Execute the query
result = g.query(query_avg_runs_by_player)

# Display average runs by player
for row in result:
    print(f"Player: {row['player']} - Average Runs: {row['avg_runs']}")
```

# Venue Analysis

```python
query_most_played_venues = """
    PREFIX onto: <http://www.semanticweb.org/user/ontologies/2024/12#>
    SELECT ?venue (COUNT(?match) AS ?matches_played)
    WHERE {
        ?match onto:hasVenue ?venue .
    }
    GROUP BY ?venue
    ORDER BY DESC(?matches_played)
    LIMIT 10
"""

# Execute the query
result = g.query(query_most_played_venues)

# Display venues with most matches played
for row in result:
    print(f"Venue: {row['venue']} - Matches Played: {row['matches_played']}")
```
